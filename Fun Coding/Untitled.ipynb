```python
##-------------------------- Optimized Code For my Code--------------------------

def isomorphic_strings(s, t):
    if len(set(s)) != len(set(t)):
        return False
    
    dict_map = {}
    
    for i, j in zip(s, t):
        if i not in dict_map:
            dict_map[i] = j
        elif dict_map[i] != j:
            return False
    
    return True

# Example usage:
s = "egg"
t = "add"
print(isomorphic_strings(s, t))  # Output: True
```
