```python
#Steming = bringing words to its root form
from nltk.stem.porter import PorterStemmer

ps= PorterStemmer()
```

```python
def stem(tags):
    y=[]
    for i in tags:
        y.append((ps.stem(i)).lower())
    return " ".join(y)
```

```python
stem(new_movies['tags'][0])
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
new_movies.head(2)
```

```python
#Convert movie's tags to vector of word's frequency

from sklearn.feature_extraction.text import CountVectorizer

cv= CountVectorizer(max_features= 5000, stop_words='english')
vectors= cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
#take the cosine similarity distance
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(vectors).shape
```

```python
similarity= cosine_similarity(vectors)
```

```python
def recommend_movies(movie):
    
    movie_id= new_movies[new_movies['original_title']== movie].index[0] 
    distance= similarity[movie_id]
    
    movie_list= sorted(list(enumerate(similarity[movie_id])), reverse=True, key= lambda x: x[1])
    
    recommended_movies=[]
    recommended_movies_poster = []
    for i in movie_list[1:6]:
        
        recommended_movies.append(new_movies['original_title'].iloc[i[0]])
        recommended_movies_poster.append(new_movies['imdb_id'].iloc[i[0]])
        
    return recommended_movies, recommended_movies_poster
    
    
    
```
